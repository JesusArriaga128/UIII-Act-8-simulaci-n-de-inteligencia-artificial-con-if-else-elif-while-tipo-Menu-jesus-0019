(Cirujano : bisturí)

```python
# Simulador de IA: Cirujano y Bisturí
# Nivel: Principiante (Solo estructuras de control básicas)

print("====================================================")
print(" ¡BIENVENIDO AL SIMULADOR DE IA: ASISTENTE DE QUIRÓFANO! ")
print("====================================================")
print("Eres el Cirujano en jefe. Tu misión es operar con éxito.")
print("Tu IA interna te guiará para usar el bisturí correctamente.")
print("====================================================")

# Variable para mantener el juego activo
ejecutando = "si"

while ejecutando == "si":
    # Mostrar el menú interactivo
    print("\n--- MENÚ DE ACCIONES EN EL QUIRÓFANO ---")
    print("1. Analizar el estado del paciente")
    print("2. Preparar el bisturí")
    print("3. Realizar incisión principal")
    print("4. Terminar la cirugía / Salir")

    opcion = input("Cirujano, elija una opción (1-4): ")

    # Evaluar la opción del usuario (Simulación de IA por reglas)
    if opcion == "1":
        print("\n[IA Analizando...] Diagnóstico: Apendicitis detectada.")
        print("Recomendación: Se requiere el uso del bisturí de inmediato.")

    elif opcion == "2":
        print("\n¿Qué tipo de bisturí deseas esterilizar?")
        print("- tradicional")
        print("- laser")
        tipo_bisturi = input("Escribe el tipo: ")

        if tipo_bisturi == "tradicional":
            print("\n[IA]: Bisturí de acero inoxidable listo y afilado.")
        elif tipo_bisturi == "laser":
            print("\n[IA]: Bisturí láser cargado al 100%. ¡Cuidado con los ojos!")
        else:
            print("\n[IA ERROR]: Ese tipo de bisturí no existe en este hospital.")

    elif opcion == "3":
        presion = input("\n¿Cuánta presión aplicarás con el bisturí? (poca / media / mucha): ")

        if presion == "poca":
            print("\n[IA]: El bisturí no cortó lo suficiente. Intenta de nuevo.")
        elif presion == "media":
            print("\n[IA Exito]: ¡Incisión perfecta! El ritmo cardíaco es estable. Gran trabajo, Cirujano.")
        elif presion == "mucha":
            print("\n[IA ALERTA]: ¡Demasiada presión! El paciente está sangrando. ¡Llamen a enfermería!")
        else:
            print("\n[IA]: El cirujano dudó y no hizo el corte.")

    elif opcion == "4":
        print("\n[IA]: Cerrando simulación. Cirugía finalizada. ¡Buen trabajo en el quirófano!")
        ejecutando = "no" # Esto rompe el ciclo while

    else:
        print("\n[IA ERROR]: Opción no válida. Por favor, concéntrese en la operación.")

print("\n====================================================")
print(" Programa terminado. ¡Gracias por jugar, Doctor! ")
print("====================================================")
```
